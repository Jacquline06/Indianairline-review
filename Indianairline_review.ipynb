```python
import re
```

```python
# Function to clean ordinal suffixes
def clean_ordinal(date_str):
    # Remove 'st', 'nd', 'rd', 'th' using regex
    return re.sub(r'(\d+)(st|nd|rd|th)', r'\1', date_str)

# Apply the cleaning function to the date column
df['cleaned_date'] = df['Date'].apply(clean_ordinal)

# Convert cleaned strings to datetime
df['parsed_date'] = pd.to_datetime(df['cleaned_date'], format='%d %B %Y')
```

```python
# Function to separate verified status and clean the review text
def clean_review(review):
    # Check if the delimiter exists
    if " | " in review:
        # Split based on the delimiter
        status, text = review.split(" | ", 1)
    else:
        status, text = "Unknown Status", review  # Default status if delimiter is absent

    # Remove unwanted characters
    text = text.replace("â€™t", "t")
    return status.strip(), text.strip()
```

```python
# Apply the function to the Review column
df[['Verified Status', 'Cleaned Review']] = df['Review'].apply(clean_review).apply(pd.Series)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model= Sequential()
# Input layer (match input shape to your data)
#model.add(Dense(256, input_dim=x_train.shape[1], activation='relu'))

# Hidden layers

model.add(Dense(5,input_dim=4,activation="relu"))
model.add(Dense(2,input_dim=4,activation="relu"))
model.add(Dense(4,input_dim=4,activation="relu"))
model.add(Dense(3,activation="softmax"))




# Compile the model for regression with 'mse' loss function
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```
